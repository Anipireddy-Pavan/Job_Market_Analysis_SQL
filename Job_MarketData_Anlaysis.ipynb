```python
#Python.
orders = [
    {"order_id": 1, "customer_id": 101, "order_date": "2024-01-01", "amount": 2500, "status": "Delivered"},
    {"order_id": 2, "customer_id": 102, "order_date": "2024-01-01", "amount": 1800, "status": "Cancelled"},
    {"order_id": 3, "customer_id": 101, "order_date": "2024-01-02", "amount": 3200, "status": "Delivered"},
    {"order_id": 4, "customer_id": 103, "order_date": "2024-01-02", "amount": 4000, "status": "Delivered"},
    {"order_id": 5, "customer_id": 104, "order_date": "2024-01-03", "amount": 1500, "status": "Returned"},
    {"order_id": 6, "customer_id": 102, "order_date": "2024-01-03", "amount": 2800, "status": "Delivered"}
]
```

```python
#Problem - 1 
#Total & Average revenue.
total_revenue = 0
delivered_orders = 0

for order in orders :
    if order['status'] == 'Delivered':
        total_revenue += order['amount']
        delivered_orders += 1

average_order_value = total_revenue / delivered_orders

print("Total Revenue:", total_revenue)
print("Average Order Value:", average_order_value)
print("Total Orders:", len(orders))



```

```python
status_count = {}

for order in orders:
    status = order['status']
    status_count[status] = status_count.get(status, 0) + 1


for k, v in status_count.items():
    print(k, ":", v)
```

```python
delivered_df = df[df['status'] == 'Delivered']

total_revenue = delivered_df['amount'].sum()
average_order_value = delivered_df['amount'].mean()
total_orders = len(df)


print(total_revenue, average_order_value, total_orders)
```

```python
daily_revenue = {}

for order in orders:
    if order['status'] == 'Delivered':
        date = order['order_date']
        daily_revenue[date] = daily_revenue.get(date, 0) + order['amount']

for date, revenue in daily_revenue.items():
    print(date, " ", revenue)
    
```

```python
customers_orders = {}

for order in orders:
    if order['status'] == 'Delivered':
        cid = order['customer_id']
        customers_orders[cid] = customers_orders.get(cid, 0) + 1

repeat_customers = [cid for cid, count in customers_orders.items() if count > 1]

print("Repeat customer", repeat_customers)
```

```python
status_breakdown = df['status'].value_counts()
print(status_breakdown)
```

```python
lost_revenue = 0
delivered_revenue = 0

for order in orders:
    if order['status'] in ['Cancelled', 'Returned']:
        lost_revenue += order['amount']
    elif order['status'] == 'Delivered':
        delivered_revenue += order['amount']

loss_percentage = (lost_revenue / delivered_revenue) * 100

print("Lost Revenue:", lost_revenue)
print("Loss Percentage:", round(loss_percentage, 2), "%")
```

```python
daily_revenue = (
    df[df['status'] == 'Delivered'].groupby('order_date')['amount'].sum()
)
print(daily_revenue)
```

```python
total_orders = len(orders)
delivered = 0
cancelled = 0
total_revenue = 0

for order in orders:
    if order['status'] == 'Delivered':
        delivered += 1
        total_revenue += order['amount']
    elif order['status'] == 'Cancelled':
        cancelled += 1

avg_order_value = total_revenue / delivered
success_rate = (delivered / total_orders) * 100
cancellation_rate = (cancelled / total_orders) * 100

print("Total Revenue:", total_revenue)
print("Average Order Value:", avg_order_value)
print("Success Rate:", round(success_rate, 2), "%")
print("Cancellation Rate:", round(cancellation_rate, 2), "%")
    
```

```python
health_score = (delivered_revenue / total_orders) * (delivered / total_orders)

print("Business Health Score:", round(health_score, 2))
```
